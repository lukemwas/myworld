```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

data=pd.DataFrame({'x':x,'y':y})

print(round(data.corr(),2))

model=smf.ols('y~x',data=data)
result=model.fit()
print(result.summary())

poly=np.poly1d(np.polyfit(x,y,3))
mypoly=np.linspace(1,22,40)

print(poly(4))

plt.scatter(x,y)
plt.plot(mypoly,poly(mypoly))
plt.show()
```

```python
from scipy import stats as st
a = [5,7,8,7,2,17,2,9,4,11,12,9,6]
b = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope,intercept,r,p,std_err=st.linregress(a,b)

def myvalue(a):
    return slope * a +intercept

myline=list(map(myvalue,a))

plt.scatter(a,b)
plt.plot(a,myline)
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

a = [5,7,8,7,2,17,2,9,4,11,12,9,6]
b = [99,86,87,88,111,86,103,87,94,78,77,85,86]

data=pd.DataFrame({'a':a,'b':b})

print(round(data.corr(),2))

mydata=smf.ols('b~a',data=data)
info=mydata.fit()
print(info.summary())

poly=np.poly1d(np.polyfit(a,b,3))
mypoly=np.linspace(2,17,15)

plt.scatter(a,b)
plt.plot(mypoly,poly(mypoly))
plt.show()
```

```python
from scipy import stats 
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

slope,intercept,r,p,std_err=stats.linregress(x,y)

def myplot(x):
    return slope * x + intercept

line=list(map(myplot,x))

plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```
