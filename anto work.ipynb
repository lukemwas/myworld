```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("survey_results_public.csv")
```

```python
data.info()
```

```python
a=data['EdLevel']
```

```python
print(a)
```

```python
a.dropna()
```

```python
acounts=a.value_counts()
acounts.plot(kind='bar',color='green')
plt.xlabel('Edlevels')
plt.ylabel('Frequency')
plt.title('LEVEL OF EDUCATION OF 71,571 PEOPLE')
plt.show()
```

```python
b=data['CodingActivities']
b.dropna()
b.info()
```

```python
plt.figure(figsize=(25,15))
bactivities=b.value_counts()
bactivities.plot(kind='bar',color='green')
plt.xlabel('CodingActivities')
plt.ylabel('No of people')
plt.title('No Of People By There CodingActivities')
plt.show()
```

```python
c=data['MainBranch']
c.dropna()
c.info()
```

```python
mainbranch=c.value_counts()
mainbranch.plot(kind='bar',color='green')
plt.xlabel('Mainbranch')
plt.ylabel('No Of People')
plt.title('No Of People By There Specialist')
plt.show()
```

```python
d=data['DevType']
d.dropna()
d.info()
```

```python
plt.figure(figsize=(15,15))
devtype=d.value_counts()
plt.pie(devtype,labels=devtype,autopct='%1.1f%%')
plt.show
```
