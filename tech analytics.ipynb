```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import csv
```

```python
with open("survey_results_public.csv") as data:
    mydata=csv.DictReader(data)
    count=Counter()
    for rec in mydata:
        myvalues=rec['EdLevel'].split(';')
        for values in myvalues:
            count[values]+=1

myvalues=[]
pple=[]

for i in count.most_common():
    myvalues.append(i[0])
    pple.append(i[1])

pple.reverse()
myvalues.reverse()
plt.bar(myvalues,pple)
plt.xticks(rotation=90)
plt.title('pple', )
```

```python
with open("survey_results_public.csv") as data:
    mydata=csv.DictReader(data)
    count=Counter()

    for rec in mydata:
        mykeys=rec['DevType'].split(';')
        for values in mykeys:
            count[values]+=1

mykeys=[]
myvalues=[]

for i in count.most_common(10):
    mykeys.append(i[0])
    myvalues.append(i[1])

plt.figure(figsize=(10,9))
plt.pie(myvalues,autopct='%1.1f%%')
plt.legend(mykeys)
plt.show()
```
